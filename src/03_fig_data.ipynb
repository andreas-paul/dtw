```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# matplotlib.use('cairo')
```

```python
data_1100 = pd.read_excel("../out/data_1100.xlsx")
data_1150 = pd.read_excel("../out/data_1150.xlsx")
```

```python
fig, axs = plt.subplots(8, sharex=True, figsize=(5,10))
plt.xticks(range(0, 290, 10), rotation=45)

# 1100
sns.lineplot(ax=axs[0], x='time_conventional', y='aragonite_%', data=data_1100, estimator=None)
sns.lineplot(ax=axs[1], x='time_conventional', y='high-magnesium-calcite_%', data=data_1100, estimator=None)
sns.lineplot(ax=axs[2], x='time_conventional', y='d13C_pl', data=data_1100, estimator=None)
sns.lineplot(ax=axs[3], x='time_conventional', y='d13C_bulk', data=data_1100, estimator=None)
sns.lineplot(ax=axs[4], x='time_conventional', y='sieve-fine-<=63ym_%', data=data_1100, estimator=None)

# 1150 
sns.lineplot(ax=axs[0], x='time_dtw_combined', y='aragonite_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[1], x='time_dtw_combined', y='high-magnesium-calcite_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[2], x='time_dtw_combined', y='d13C_pl', data=data_1150, estimator=None)
sns.lineplot(ax=axs[3], x='time_dtw_combined', y='d13C_bulk', data=data_1150, estimator=None)
sns.lineplot(ax=axs[4], x='time_dtw_combined', y='sieve-<63ym_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[5], x='time_dtw_combined', y='organic-carbon_thermo-gravimetry_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[6], x='time_dtw_combined', y='total-organic-carbon_thermo-gravimetry_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[6], x='time_dtw_combined', y= 'total-organic-carbon_Flash-EA1112_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[7], x='time_dtw_combined', y='calcium-carbonate_thermo-gravimetry_%', data=data_1150, estimator=None)
sns.lineplot(ax=axs[7], x='time_dtw_combined', y='calcium-carbonate_Flash-EA1112_%', data=data_1150, estimator=None)

for ax in axs:
    ax.set_yticklabels(ax.get_yticklabels())
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha("right")
```
